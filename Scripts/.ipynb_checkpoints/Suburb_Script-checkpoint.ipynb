```python
# Splitting the HTML string based on "option value" and extracting suburb names
suburb_names = []
for line in html_string.split('<option value="')[1:]:
    if '">' in line:
        suburb = line.split('">')[1].split('</option')[0]
        suburb_names.append(suburb)

# Create the new suburb encoding dictionary
suburb_encoding = {suburb: index for index, suburb in enumerate(suburb_names)}
len(suburb_encoding), list(suburb_encoding.items())[:5]  # Displaying the length and first 5 items for verification
```

```python
# Convert the dictionary to a DataFrame
df_suburb_encoding = pd.DataFrame(list(suburb_encoding.items()), columns=['Suburb', 'Encoding'])

# Save the DataFrame as a CSV
df_suburb_encoding.to_csv('../Cleaned_Data/Suburb_List.csv', index=False)
```
